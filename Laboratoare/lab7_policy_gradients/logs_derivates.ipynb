```python
import numpy as np
import matplotlib.pyplot as plt

p = np.linspace(0.001, 0.999, 200)
y = 1

loss_no_log = np.abs(p - y)
loss_log = -np.log(p)

plt.plot(p, loss_no_log, label="|p - y| (fără log)")
plt.plot(p, loss_log, label="-log(p) (cu log)")
plt.legend()
plt.title("Fără log vs Cu log")
plt.xlabel("p")
plt.ylabel("Loss")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Target
x = 2
y = 1

# Inițializare w
w = 0.1

lr = 0.1
epochs = 50

history_no_log = []
history_log = []

# ----------------------
# 1) Fără log
# ----------------------
w_no_log = w

for _ in range(epochs):
    p = sigmoid(w_no_log * x)
    loss = abs(p - y)
    history_no_log.append(loss)

    # derivata loss = sign(p-y) * p*(1-p)*x
    grad = np.sign(p - y) * p * (1 - p) * x
    w_no_log -= lr * grad

# ----------------------
# 2) Cu log
# ----------------------
w_log = w

for _ in range(epochs):
    p = sigmoid(w_log * x)
    loss = -np.log(p)
    history_log.append(loss)

    # derivata loss = -(1/p) * p*(1-p)*x = -(1-p)*x
    grad = -(1 - p) * x
    w_log -= lr * grad

# ----------------------
plt.plot(history_no_log, label="Fără log (|p-y|)")
plt.plot(history_log, label="Cu log (-log(p))")
plt.legend()
plt.title("Învață rețeaua? Cu log vs fără log")
plt.xlabel("Epoca")
plt.ylabel("Loss")
plt.grid(True)
plt.show()
```
